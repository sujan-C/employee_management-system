<h2>Step 5. Data Visualization </h2>

```python
# Create a cursor object to execute SQL queries
cursor = connection.cursor()
# Query data from the 'employees' table
cursor.execute('SELECT DepartmentID, COUNT(*) as EmployeeCount FROM employees GROUP BY DepartmentID')
#After fetching data from the database we are storing it into Pandas DataFrame
employeesCount_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
# Print the first few rows of the DataFrame to verify the data
print(employeesCount_df.head())
# Plot the graph
plt.figure(figsize=(12, 8))
plt.bar(employeesCount_df['DepartmentID'], employeesCount_df['EmployeeCount'])

plt.xlabel('Department ID')
plt.ylabel('Number of Employees')
plt.title('Number of Employees per Department')
plt.xticks(employeesCount_df['DepartmentID'])
plt.show()
```

```python


# Assuming you have a connection object already created, use it directly
# connection = ...

# Write the SQL query to retrieve the required data
query = "SELECT HireDate FROM employees"

# Execute the query and load the data into a DataFrame
df = pd.read_sql(query, connection)

# Convert the HireDate column to datetime
df['HireDate'] = pd.to_datetime(df['HireDate'])

# Count the number of employees hired over time and calculate the cumulative sum
hire_counts = df['HireDate'].value_counts().sort_index().cumsum()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(hire_counts.index, hire_counts.values, marker='o', linestyle='-', color='b')
plt.xlabel('Hire Date')
plt.ylabel('Cumulative Number of Employees Hired')
plt.title('Number of Employees Hired Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

# Write the SQL query to retrieve the required data
query = "SELECT departmentname, budget FROM departments LIMIT 20"

# Execute the query and load the data into a DataFrame
df = pd.read_sql(query, connection)

# Extract department names and budgets
departments = df['departmentname'].tolist()
budgets = df['budget'].tolist()

# Plotting the pie chart
plt.figure(figsize=(12, 8))
plt.pie(budgets, labels=departments, autopct='%1.1f%%', startangle=140)
plt.title('Department Budgets Distribution (First 20 Records)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

# Write the SQL query to retrieve the required data
query = "SELECT managerid, budget FROM departments limit 20"

# Execute the query and load the data into a DataFrame
df = pd.read_sql(query, connection)

# Extract manager_ids and budgets from the DataFrame
manager_ids = df['managerid'].tolist()
budgets = df['budget'].tolist()

# Plotting the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(manager_ids, budgets, color='red')
plt.xlabel('Manager ID')
plt.ylabel('Budget')
plt.title('Manager ID vs Budget')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming you have a connection object already created, use it directly
# connection = ...

# Write the SQL query to retrieve the required data
query = "SELECT EmployeeID, hoursworked FROM attendance limit 20"

# Execute the query and load the data into a DataFrame
df = pd.read_sql(query, connection)

# Convert hours_worked to a list
hours_worked = df['hoursworked'].tolist()
employee_ids = df['EmployeeID'].tolist()

# Create polar area chart
angles = np.linspace(0, 2 * np.pi, len(employee_ids), endpoint=False).tolist()
hours_worked += hours_worked[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))
ax.fill(angles, hours_worked, color='teal', alpha=0.25)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(employee_ids)
ax.set_title('Polar Area Chart of Hours Worked by Employees')
plt.show()
```
